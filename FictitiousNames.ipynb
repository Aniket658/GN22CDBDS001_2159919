### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Create the 3 DataFrames based on the following raw data

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

### Step 3. Assign each to a variable called data1, data2, data3

```python
data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)
data3 = pd.DataFrame(raw_data_3)
# Converting Dictionary to DataFrame
```

```python
data1.head()
```

```python
data2.head()
```

```python
data3.head(10)
```

### Step 4. Join the two dataframes along rows and assign all_data

```python
all_data = pd.concat([data1, data2]) # Concating Data1 and Data2 along row
all_data.head(10)
```

### Step 5. Join the two dataframes along columns and assing to all_data_col

```python
all_data_col = pd.merge(data1, data3) # Merging Data1 and Data3 along Column (It uses subject_id as common column by default).
all_data_col.head(10)
```

### Step 6. Print data3

```python
print(data3)
```

### Step 7. Merge all_data and data3 along the subject_id value

```python
pd.merge(all_data, data3, on="subject_id")
# Merging all_data and Data3 along Column by subject_id
```

### Step 8. Merge only the data that has the same 'subject_id' on both data1 and data2

```python
pd.merge(data1, data2, on="subject_id")
# Merging Data1 and Data2 along Column by subject_id
```

### Step 9. Merge all values in data1 and data2, with matching records from both sides where available.

```python
pd.merge(data1, data2, on='subject_id', how = 'outer')
# Using outer as we want to display all rows from both dataframe.
```
